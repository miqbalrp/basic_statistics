```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
import researchpy as rp

%matplotlib inline
```

## Import Data

```python
data_dir = r'data//blood_pressure.csv'

df = pd.read_csv(data_dir)
df.info()
```

# T-Test

```python
group1= df['bp_after'][df['sex'] == 'Male']
group2= df['bp_after'][df['sex'] == 'Female']
```

## 1. Visualize Data

```python
fig_title = 'Data Visualization'
fig, axs = plt.subplots(1,2, figsize=(12,5))
fig.suptitle(fig_title, fontsize=15)

sns.countplot(ax=axs[0], data=df, x='sex', edgecolor='k')
sns.boxplot(ax=axs[1], data=df, y='sex', x='bp_after')
plt.show()
```

## 2. Normality and Variance Equality Assumption

### a. Normality Check

```python
alpha = 0.01

print('Null hypothesis         : Population normally distributed')
print('Alternative hypothesis  : Population not normally distributed')
print('Significant level       : {:}'.format(alpha))

fig_title = 'Q-Q Plot ($\\alpha$={:})'.format(alpha)
fig, axs = plt.subplots(1,2, figsize=(12,5))
fig.suptitle(fig_title, fontsize=15)
pg.qqplot(group1, dist='norm', confidence=1-alpha, ax=axs[0])
pg.qqplot(group2, dist='norm', confidence=1-alpha, ax=axs[1])
axs[0].title.set_text('Male')
axs[1].title.set_text('Female')
plt.show()

pg.normality(df, dv='bp_after', group='sex', alpha=0.01)
```

### b. Equal Variance Check

```python
alpha = 0.01

print('Null hypothesis         : All variances are equal')
print('Alternative hypothesis  : At least one variance is different')
print('Significant level       : {:}'.format(alpha))

pg.homoscedasticity(data=df, dv='bp_after', group='sex', alpha=alpha)
```

* Using Q-Q Plot and Shapiro-Wilk test with 99% confidence, we get p-value of 0.355 and 0.014 which means that we cannot reject the null hypotesis (the population is normally distributed)
* Using Levene test, we get p-value 0.026 for equal variance test. So we cannot reject the null hypothesis which is two population have equal variance.
* In conclusion, we can use Student's t-test.

## 4. Define H0 and H1

```python
print('Null hypothesis         : Blood pressure of male and female after treatment are equal')
print('Alternative hypothesis  : Blood pressure of male and female after treatment are different')
print('Test                    : t-test with two-sided tail')
```

## 5. Conduct T-test and Conclusion

```python
alpha = 0.05
result = pg.ttest(group1,group2, paired=False, tail='two-sided', confidence=1-alpha)
result['diff'] = np.mean(group1) - np.mean(group2)
pd.melt(result)
```

* From result above, we get that p-value < alpha. The null hypothesis is rejected and there is significant difference of male and female. 
* We 95% confidence that difference is between 3.4 to 13.24.

# ANOVA

## 1. Visualization

```python
fig_title = 'Data Visualization'
fig, axs = plt.subplots(1,2, figsize=(12,5))
fig.suptitle(fig_title, fontsize=15)

sns.countplot(ax=axs[0], data=df, x='agegrp', edgecolor='k')
sns.boxplot(ax=axs[1], data=df, y='agegrp', x='bp_after')
plt.show()
```

## 2. Normality and Equal Variance Assumptions

```python
group1= df['bp_after'][df['agegrp'] == '30-45']
group2= df['bp_after'][df['agegrp'] == '46-59']
group3= df['bp_after'][df['agegrp'] == '60+']
```

### a. Normality

```python
alpha = 0.05

print('Null hypothesis         : Population normally distributed')
print('Alternative hypothesis  : Population not normally distributed')
print('Significant level       : {:}'.format(alpha))

fig_title = 'Q-Q Plot ($\\alpha$={:})'.format(alpha)
fig, axs = plt.subplots(1,3, figsize=(17,5))
fig.suptitle(fig_title, fontsize=15)
pg.qqplot(group1, dist='norm', confidence=1-alpha, ax=axs[0])
pg.qqplot(group2, dist='norm', confidence=1-alpha, ax=axs[1])
pg.qqplot(group3, dist='norm', confidence=1-alpha, ax=axs[2])
axs[0].title.set_text('30-45')
axs[1].title.set_text('45-59')
axs[2].title.set_text('60+')
plt.show()

pg.normality(df, dv='bp_after', group='agegrp', alpha=0.01)
```

### b. Equal Variance

```python
alpha = 0.01

print('Null hypothesis         : All variances are equal')
print('Alternative hypothesis  : At least one variance is different')
print('Significant level       : {:}'.format(alpha))

pg.homoscedasticity(data=df, dv='bp_after', group='agegrp', alpha=alpha)
```

## 4. Define H0 and H1

```python
print('Null hypothesis         : Blood pressure mean of each age group are equal')
print('Alternative hypothesis  : Blood pressure mean of each age group are different')
print('Test                    : ANOVA')
```

## 5. ANOVA between One Factor (One-Way ANOVA)

```python
alpha = 0.05
result = pg.anova(data=df, dv='bp_after', between='agegrp', detailed=False)
result
```

## 6. ANOVA between Two Factors (Two-Way ANOVA)

```python
alpha = 0.05
result = pg.anova(data=df, dv='bp_after', between=['agegrp', 'sex'], detailed=False)
result
```
